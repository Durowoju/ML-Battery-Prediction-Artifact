```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
from joblib import dump
import os


# Load the data from the Excel file
data = pd.read_excel("C:/Users/Emmanuel Durowoju/Desktop/Impedance/July03/25C/Li3InCl6F_Bat_25_EIS_4A.xlsx", sheet_name=3)

# Extract features (x) and target variable (y)
y = data[['Z1(b)']].values.astype(float).reshape(-1)
x = data[['Z1(a)']].values.astype(float)

# Create a Random Forest Regression model
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=0)
rf_regressor.fit(x, y)
dump(rf_regressor, 'RF_Model1.joblib')

# Predict the values
x_pred = np.linspace(min(x), max(x), 100).reshape(-1, 1)
y_pred = rf_regressor.predict(x_pred)

# Plot the original data and the regression line
plt.scatter(x, y, label='Data')
plt.plot(x_pred, y_pred, color='red', label='Random Forest Regression')
plt.xlabel('Z1(a)')
plt.ylabel('Z1(b)')
plt.legend()

# Find x values when y equals zero
x1 = x_pred[np.argmin(np.abs(y_pred))]
x2 = x_pred[np.argmax(np.abs(y_pred))]
#final_model.save_modek('seg.model')

# Plot the points where y is zero
plt.scatter([x1, x2], [0, 0], color='green', label='y=0 points')
plt.legend()

# Draw horizontal line at y=0
plt.axhline(0, color='gray', linestyle='--', label='y=0')

# Annotate the plot with x1 and y values
for i in range(0, len(x_pred), 5):  # Annotate every 10th point
    x1_new = x_pred[i][0]
    y_new = y_pred[i]
    plt.annotate(f'{x1_new:.2f}', xy=(x1_new, y_new), ha='left', va='bottom', rotation=45, fontsize=8)

# Show the plot
plt.show()
```
